## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df["subtotal"] = df["unit_price"] * df["qty"]
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
# from previous calculation, no unit_weight exceeds 46.43 pounds, so I need to calculate total order weight with qty and unit_weight, since everything is under 50 pounds. In my opinion this wording should have been clearer, since items is not clear at all
def shipping_price(row):
    order_weight = row["unit_weight"] * row["qty"]
    if order_weight > 50:
        return 7 * order_weight
    else:
        return 10 * order_weight

df["shipping_price"] = df.apply(shipping_price, axis=1)
df.head(15)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
def total_price(row):
    price = round(((row["subtotal"] + row["shipping_price"]) * 1.0925),2)
    return price

df["total_price"] = df.apply(total_price, axis=1)
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df["line_cost"] = df["unit_cost"] * df["qty"] + df["shipping_price"]
df.head()
```

```python
# Create a column for the profit of each line using line cost and line price
df["profit"] = df["total_price"] - df["line_cost"]
df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
for x in top_five_clients:
    total_spent = round(df[df['client_id']==x]['total_price'].sum(),2)
    client_name = df[df['client_id']==x]["first"].values[0] + " " + df[df['client_id']==x]["last"].values[0]
    print(f"Client {x}, who is {client_name}, spent a total of ${total_spent:,}")
    total_spent=0 # reset so that I don't mess things up for the next calculation
```

```python
df.columns
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
total_units_purchased = []
total_shipping_price = []
total_revenue = []
total_profit = []
client_names = []

#loop through the client list and append to all the lists for the dataframe
for x in top_five_clients:
    total_units_purchased.append(df[df['client_id']==x]['qty'].sum())
    total_shipping_price.append(df[df['client_id']==x]['shipping_price'].sum())
    total_revenue.append(df[df['client_id']==x]['total_price'].sum())
    total_profit.append(df[df['client_id']==x]['profit'].sum())
    client_names.append(df[df['client_id']==x]["first"].values[0] + " " + df[df['client_id']==x]["last"].values[0])

# summary dataframe using the created lists
summary_top_five_df = pd.DataFrame({"Client Name" : client_names, "Client ID": top_five_clients, "Total Units Purchased": total_units_purchased, "Total Shipping Price": total_shipping_price, "Total Revenue": total_revenue, "Total Profit":total_profit})
summary_top_five_df.head()
```

```python
# Format the data and rename the columns to names suitable for presentation.
# I believe I already did this when I made the dataframe?

# Define the money columns. 
money_columns = ['Total Shipping Price', 'Total Revenue', 'Total Profit']

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return x/1000000

# Apply the currency_format_millions function to only the money columns. 
summary_top_five_df[money_columns] = summary_top_five_df[money_columns].apply(currency_format_millions)
# Rename the columns to reflect the change in the money format. 
summary_top_five_df= summary_top_five_df.rename(columns={"Total Shipping Price": "Total Shipping Price (Millions USD)", "Total Revenue": "Total Revenue (Millions USD)", "Total Profit" : "Total Profit (Millions USD)"})
summary_top_five_df.head()

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_five_clients = summary_top_five_df.sort_values(by="Total Profit (Millions USD)", ascending=False)
sorted_five_clients = sorted_five_clients.reset_index(drop=True)
sorted_five_clients.head()
```
